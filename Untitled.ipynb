### Let's install some libraries and import our data

```python
# Import our data and install library
!pip install plotly
import pandas as pd
import altair as alt
import plotly.graph_objects as go
import numpy as np


# Our Dataset about UEFA Euro 2024 players
data= pd.read_csv("euro2024_players.csv")
data.head()
```

### Box plot for each national team

```python
# Box plots for each national team, displaying the distribution of player market values
box_plot = alt.Chart(data).mark_boxplot().encode(
    x='Country:N',
    y='MarketValue:Q',
    color='Country:N'
).properties(
    title='Distribution of Player Market Values by National Team'
)

box_plot
```

### Heatmap with Player Positions and Age Ranges

```python
# Define age ranges
data['AgeRange'] = pd.cut(data['Age'], bins=[15, 20, 25, 30, 35, 40, 45], labels=['15-20', '21-25', '26-30', '31-35', '36-40', '41-45'])

# Heatmap with player positions on one axis and age ranges on the other
heatmap = alt.Chart(data).mark_rect().encode(
    x='Position:N',
    y='AgeRange:N',
    color='count():Q'
).properties(
    title='Number of Players by Position and Age Range'
)

heatmap
```

### Chord Diagram to Illustrate the Connections Between Clubs and National Teams

Unfortunately, Altair does not natively support chord diagrams. However, we can create a similar visualization using a matrix plot or use another library such as Plotly.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Prepare data
club_country = data.groupby(['Club', 'Country']).size().reset_index(name='count')

# Create a graph
G = nx.Graph()

# Add nodes
clubs = data['Club'].unique()
countries = data['Country'].unique()
G.add_nodes_from(clubs, bipartite=0)
G.add_nodes_from(countries, bipartite=1)

# Add edges
edges = [(row['Club'], row['Country'], row['count']) for _, row in club_country.iterrows()]
G.add_weighted_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G, k=0.3)
plt.figure(figsize=(14, 14))

# Draw nodes
nx.draw_networkx_nodes(G, pos, nodelist=clubs, node_color='r', node_size=50, alpha=0.8, label='Club')
nx.draw_networkx_nodes(G, pos, nodelist=countries, node_color='b', node_size=50, alpha=0.8, label='National Team')

# Draw edges
edges = nx.draw_networkx_edges(G, pos, alpha=0.3)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=8)

plt.legend(scatterpoints=1)
plt.title('Connections Between Clubs and National Teams')
plt.axis('off')
plt.show()
```

```python
# Prepare data
club_country = data.groupby(['Club', 'Country']).size().reset_index(name='count')

# Filter to top 10 clubs with most players
top_clubs = data['Club'].value_counts().nlargest(10).index
filtered_data = data[data['Club'].isin(top_clubs)]

# Update the club_country for the filtered data
club_country_filtered = filtered_data.groupby(['Club', 'Country']).size().reset_index(name='count')

# Create a graph
G = nx.Graph()

# Add nodes
clubs = filtered_data['Club'].unique()
countries = filtered_data['Country'].unique()
G.add_nodes_from(clubs, bipartite=0)
G.add_nodes_from(countries, bipartite=1)

# Add edges using the filtered data
edges = [(row['Club'], row['Country'], row['count']) for _, row in club_country_filtered.iterrows()]
G.add_weighted_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G, k=0.3)
plt.figure(figsize=(14, 14))

# Draw nodes
nx.draw_networkx_nodes(G, pos, nodelist=clubs, node_color='r', node_size=100, alpha=0.8, label='Club')
nx.draw_networkx_nodes(G, pos, nodelist=countries, node_color='b', node_size=100, alpha=0.8, label='National Team')

# Draw edges
edges = nx.draw_networkx_edges(G, pos, alpha=0.5)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=8)

plt.legend(scatterpoints=1)
plt.title('Connections Between Top Clubs and National Teams')
plt.axis('off')
plt.show()

# Count the number of unique clubs
unique_clubs = filtered_data['Club'].nunique()
print(f'There are {unique_clubs} unique clubs.')

# Count the number of players in each club
players_per_club = filtered_data['Club'].value_counts()
print(players_per_club)
```

```python
# Assuming 'filtered_data' is the dataframe that contains only the top clubs

# Step 1: Group by Country and Club to get the count of players per club per country
club_country_counts = filtered_data.groupby(['Country', 'Club']).size().reset_index(name='player_count')

# Step 2: Group by Country to get the total number of players each national team receives from the top clubs
national_team_players = club_country_counts.groupby('Country')['player_count'].sum().reset_index()

# Step 3: Sort the results from lowest to highest
national_team_players_sorted = national_team_players.sort_values(by='player_count', ascending=False)

# Display the results without the index
print(national_team_players_sorted.to_string(index=False))

```
